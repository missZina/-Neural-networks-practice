# Example 3

```
from skimage.filters import gabor
from skimage import data, io
from matplotlib import pyplot as plt  
image = data.clock()
plt.imshow(image)

```

```
# detecting edges in a clock image
filt_real, filt_imag = gabor(image, frequency=0.6)
plt.figure()            
io.imshow(filt_real)    
io.show()               
# less sensitivity to finer details with the lower frequency kernel 0.1
filt_real, filt_imag = gabor(image, frequency=0.1)
plt.figure()            
io.imshow(filt_real)    
io.show() 
 # less sensitivity to finer details with the superior frequency kernel 1
filt_real, filt_imag = gabor(image, frequency=1)
plt.figure()            
io.imshow(filt_real)    
io.show()


```

# Example 4

```
# Edges filtering
import ee
from skimage import io
from matplotlib import pyplot as plt
image=data.horse()     
from skimage.filters import roberts, sobel ,scharr , prewitt

edg_roberts = roberts(image)
plt.figure() 
io.imshow(edg_roberts)    
io.show()
edg_sobel = sobel(image)
plt.figure() 
io.imshow(edg_sobel)    
io.show()
edg_scharr = scharr(image)
plt.figure() 
io.imshow(edg_scharr)    
io.show()
edg_prewitt = prewitt(image)
plt.figure() 
io.imshow(edg_prewitt)    
io.show()
from skimage.feature import canny
edg_canny = canny(image)
plt.imshow(edg_canny)
```
